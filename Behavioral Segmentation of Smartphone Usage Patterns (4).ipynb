## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_data = sc.fit_transform(df)

scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
scaled_df.head()
```

## ðŸ“ Feature Scaling using StandardScaler

K-Means clustering is distance-based and relies on Euclidean distance for grouping data points.

Since features in the dataset have different scales (e.g., Phone_Checks_Per_Day vs Sleep_Hours), feature scaling is required to ensure that no variable dominates the distance calculation.

StandardScaler was applied to:
- Normalize all features
- Transform data to have mean = 0 and standard deviation = 1

This ensures fair contribution of each feature in cluster formation.


## K-Means Clustering

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i, init='k-means++', random_state=42)
    km.fit(scaled_df)
    wcss.append(km.inertia_)
    
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.xticks(range(1, 11))
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

## ðŸ¤– K-Means Clustering Algorithm

K-Means is an unsupervised machine learning algorithm used for grouping similar data points into clusters.

The algorithm works as follows:

1. Select number of clusters (k).
2. Initialize k centroids randomly.
3. Assign each data point to the nearest centroid.
4. Recalculate centroids based on cluster means.
5. Repeat steps 3â€“4 until convergence.

K-Means aims to minimize the Within-Cluster Sum of Squares (WCSS), ensuring that data points within each cluster are as similar as possible.


## ðŸ“‰ Elbow Method for Optimal Cluster Selection

The Elbow Method was used to determine the optimal number of clusters.

WCSS (Within-Cluster Sum of Squares) was calculated for multiple values of k. As k increases, WCSS decreases.

The optimal number of clusters is selected at the "elbow point" where the rate of decrease sharply changes.

This indicates diminishing returns in compactness improvement.


## SILHOUETTE SCORE

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []

for i in range(2, 11):  # start from 2 because silhouette needs at least 2 clusters
    kmeans = KMeans(n_clusters=i, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_df)
    score = silhouette_score(scaled_df, cluster_labels)
    silhouette_scores.append(score)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score vs Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

## ðŸ“Š Silhouette Score Validation

Silhouette Score measures how well data points fit within their assigned cluster compared to other clusters.

The score ranges from -1 to +1:

- +1 â†’ Strong clustering
- 0 â†’ Overlapping clusters
- -1 â†’ Incorrect clustering

The number of clusters that maximized the silhouette score was selected for the final model.


## APPLY FINAL KMEANS & ATTACH CLUSTERS
